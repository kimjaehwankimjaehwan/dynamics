```matlab
%  
v0 = 40;   %   (m/s)
y0 = 1.2;  %   (m)
g = 9.81;  %   (m/s^2)
x_B = 10;  % A B   (m)

% a) B  h 
t_B = x_B / v0;  % B 
h_B = y0 - (1/2) * g * t_B^2;  % B 

% b) B  alpha 
vy_B = -g * t_B;  % B  
vx_B = v0;        %   
alpha_B = atan2(vy_B, vx_B) * 180 / pi;  %  (  )

%  
fprintf('B  h: %.2f m\n', h_B);
```

```matlab
%   
syms t

%   
x = 2*t^3 - t^2 + t;  % x(t) in mm
y = 2*t^2 - t;        % y(t) in mm

%    ( 1 )
vx = diff(x, t);
vy = diff(y, t);

%    ( 1 )
ax = diff(vx, t);
ay = diff(vy, t);

%   
t_vals = linspace(0, 2, 100);  % t = 0 2
x_vals = double(subs(x, t, t_vals));  % x(t) 
y_vals = double(subs(y, t, t_vals));  % y(t) 

%     
vx_vals = double(subs(vx, t, t_vals));  % vx(t)
vy_vals = double(subs(vy, t, t_vals));  % vy(t)
ax_vals = double(subs(ax, t, t_vals));  % ax(t)
ay_vals = double(subs(ay, t, t_vals));  % ay(t)

%     
v_vals = sqrt(vx_vals.^2 + vy_vals.^2);  %  
a_vals = sqrt(ax_vals.^2 + ay_vals.^2);  %  
theta_vals = atan2(vy_vals, vx_vals) * 180 / pi;  %  ( )

%   
figure;
h_traj = plot(x_vals, y_vals, 'k--', 'LineWidth', 1.5);  % 
hold on;
h_point = plot(x_vals(1), y_vals(1), 'ro', 'MarkerSize', 8, 'MarkerFaceColor', 'r');  %   

% ,    ()
h_velocity = quiver(x_vals(1), y_vals(1), vx_vals(1), vy_vals(1), 'b', 'LineWidth', 2);
h_acceleration = quiver(x_vals(1), y_vals(1), ax_vals(1), ay_vals(1), 'g', 'LineWidth', 2);

%   ,    
info_text = text(x_vals(1), y_vals(1)+5, sprintf('v: %.2f mm/s, a: %.2f mm/s^2, : %.2f', v_vals(1), a_vals(1), theta_vals(1)), 'FontSize', 12, 'Color', 'k');

%   
xlabel('X (mm)');
ylabel('Y (mm)');
title(' : ,    tan() ');
grid on;
xlim([min(x_vals)-10, max(x_vals)+10]);
ylim([min(y_vals)-10, max(y_vals)+10]);

%  
for i = 1:length(t_vals)
    %   
    set(h_point, 'XData', x_vals(i), 'YData', y_vals(i));

    %   
    set(h_velocity, 'XData', x_vals(i), 'YData', y_vals(i), ...
        'UData', vx_vals(i), 'VData', vy_vals(i));  %  

    %   
    set(h_acceleration, 'XData', x_vals(i), 'YData', y_vals(i), ...
        'UData', ax_vals(i), 'VData', ay_vals(i));  %  

    %   (, ,  )
    set(info_text, 'Position', [x_vals(i), y_vals(i)+5], ...
        'String', sprintf('v: %.2f mm/s, a: %.2f mm/s^2, : %.2f', v_vals(i), a_vals(i), theta_vals(i)));

    %  
    drawnow;

    pause(0.05);  %   
end

hold off;
```
